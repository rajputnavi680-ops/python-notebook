```python
def pass_student(**kwargs):
  pass_student_list = []
  for name, marks in kwargs.items():
    if marks>35:
      pass_student_list.append(name)
  return pass_student_list
pass_student(A=10, B= 20,C= 40, D=50, E = 60, F = 30)
```

super vised learning -> it is a type of machine learning that deal with label data it solves two types of problems
1 regression based oproblems
2 classification based problem

```python
#  label data -> theres are two types of feature
#  1 -> input feature/independent feaature eg package predication cgpa -> 10 7 8
# 2-> dependent feature / target value/ output feature eg package 9.5 6.5 7.2
```

```python
# regration based problems -> in this problem the data type of output feature is continous.
# 1 package predication
#  input feat.     output feat.
# cgpa             package
  #  10               9.5
  #  9                8
  #  8                7.5
# 2 house price prediction
#  I/p              o/p
# no of room         price
#  8                    15
#  6                    10
```

```python
#  classification based problem -> in this problem the output feature data types is discrete form
#  eg -> result prediction,spam mail prediction
#  marks             result
  # 70                 pass->1
  # 32                  fail->0
  # 65                  pass->1
```

cosine similarity


```python
import numpy as np
from numpy.linalg import norm
```

```python
A = np.array([2,1,3,2,4])
B = np.array([3,4,2,5,5])
# compute cosine similarty
cosine = np.dot(A,B)/norm(A)*norm(B)
print("cosine similarity:",cosine)
```
